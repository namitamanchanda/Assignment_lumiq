```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('phase_data_sample.csv')
data.head()
```

```python
data.label.unique()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(),cmap='rainbow')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=400)
```

```python
---Logistic regression----
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
sns.heatmap(confusion_matrix(y_test,y_pred),cmap='vlag',annot=True,fmt='.3g')
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(criterion='gini')
rfc.fit(x_train,y_train)
```

```python
y1_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
sns.heatmap(confusion_matrix(y_test,y1_pred),cmap='gist_earth',annot=True,fmt='.3g')
```

```python
print(accuracy_score(y_test,y1_pred))
```

```python
print(classification_report(y_test,y1_pred))
```

```python
from sklearn.naive_bayes  import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y2_pred=gnb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
sns.heatmap(confusion_matrix(y_test,y2_pred),cmap='copper',annot=True,fmt='.3g')
```

```python
print(accuracy_score(y_test,y2_pred))
```

```python
print(classification_report(y_test,y2_pred))
```
